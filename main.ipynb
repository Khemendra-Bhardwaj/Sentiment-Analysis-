```python
import numpy as np
import pandas as pd
import re
```

```python
data= pd.read_csv("youtube_dataset.csv")

data.head(10)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data["Comment"] = data["Comment"].str.replace("[^a-zA-Z#]", " ")
data["Comment"] = data["Comment"].apply(lambda x: ' '.join([w for w in x.split() if len(w)>2]))
data["Comment"] = data["Comment"].apply(lambda x:x.lower())
```

```python
data.head(10)
```

```python
data2=data.sample(10000)
data2.shape
```

```python
import nltk
```

```python
data2['Comment'] = data['Comment'].apply(lambda x: x.split())
data2.head()
```

```python
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

wnl = WordNetLemmatizer()
```

```python
data2['Comment'].apply(lambda x: [wnl.lemmatize(i) for i in x if i not in set(stopwords.words('english'))])
data2.head()
```

```python
def join_back(list_input):
  return " ".join(list_input)

data2['Comment'] =data2['Comment'].apply(join_back)
data2.shape
data2.head(10)
```

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
data2['Sentiment Scores'] = data2['Comment'].apply(lambda x:sia.polarity_scores(x)['compound'])
```

```python
data2['Sentiment'] = data2['Sentiment Scores'].apply(lambda s : 'Positive' if s > 0 else ('Neutral' if s == 0 else 'Negative'))
data2.head(10)
```

```python
data2.drop(['Sentiment Scores'], axis=1,inplace=True)
data2.drop(['Date','Likes'],axis=1,inplace=True)
data2.head(5)
```

Dataframe is now ready

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000)
```

```python
X= cv.fit_transform(data2['Comment']).toarray()
```

```python
X.shape
```

```python
data2['Sentiment'].replace({'Positive':1,'Negative':0,'Neutral':2},inplace=True)
data2.head(3)
```

```python
Y=data2.iloc[:,-1].values
```

```python
Y.shape
```

Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB

nb1=GaussianNB()
nb2=MultinomialNB()
nb3=BernoulliNB()
```

```python
nb1.fit(x_train,y_train)
nb2.fit(x_train,y_train)
nb3.fit(x_train,y_train)
```

```python
y_pred1=nb1.predict(x_test)
y_pred2=nb2.predict(x_test)
y_pred3=nb3.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Gaussian/Normal distribution Naive Bayes accuracy",accuracy_score(y_test,y_pred1))

print("Multinomial distribution Naive Bayes accuracy",accuracy_score(y_test,y_pred2))

print("Bernouli distribution Naive Bayes accuracy",accuracy_score(y_test,y_pred3))
```
